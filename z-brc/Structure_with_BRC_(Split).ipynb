```python
import pandas as pd

# fetch dataset
brca = pd.read_csv("/content/drive/My Drive/Datasets/brc.csv")
brca.drop(columns=['Unnamed: 32'], inplace = True)


# data (as pandas dataframes)
X = brca.drop(columns=['id', 'diagnosis'])
y = brca['diagnosis']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
```

```python
# Concatenate the two DataFrames
brc_dr_train = X_train.join(y_train)
brc_dr_test = X_test.join(y_test)
```
